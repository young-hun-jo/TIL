```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup
```

```python
chromedriver = "C:/webdriver/chrome/chromedriver.exe"
driver = webdriver.Chrome(chromedriver)
driver.get("https://www.youtube.com/watch?v=ZnceJMyZyAs")

time.sleep(3)

body = driver.find_element_by_tag_name('body')

num_of_pagedowns = 10

while num_of_pagedowns:
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.7)
    num_of_pagedowns -= 1

html_source = driver.page_source
    
driver.quit()

soup = BeautifulSoup(html_source, 'html.parser')
comments = soup.select("yt-formatted-string#content-text")
ids = soup.select("div#header-author > a#author-text > span")
for index, i in enumerate(range(len(comments))):
    id = ids[i].text.strip("\n").strip()
    comment = comments[i].text.strip("\n").strip()
    print(f"{index+1}. {id}: {comment}")
    
```
